I declared a function which will quickly plot bar charts, to save me time everytime I want to plot one. 

```python
def plot_cluster_bar_chart(data,title,xlabel):
    plt.figure(figsize=(7,5))
    y_pos = np.arange(len(data))
    labels = data
    colors=['blue', 'darkorange', 'green']
    cluster_numbers = ['Cluster 1','Cluster 2','Cluster 3']
    plt.barh(y_pos,data,color=colors)
    plt.yticks(y_pos, cluster_numbers)
    plt.xlabel(xlabel)
    plt.title(title)
    plt.show()
```

```python
correlation_df = df[['hashedathleteid','cluster','average_weekly_trainings','average_weekly_training_hours','average_weekly_totaldistance','average_speed','marathon_finish_time']]
correlation_df = correlation_df[~correlation_df.hashedathleteid.duplicated(keep='first')]
correlation_df = correlation_df.reset_index(drop=True)
```
